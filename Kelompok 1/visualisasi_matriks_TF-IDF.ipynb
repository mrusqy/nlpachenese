## Memvisualisasikan matriks TF-IDF:

```python
import pandas as pd

# Memuat data dari file CSV
tfidf_df = pd.read_csv('tfidf_output.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengambil subset dari DataFrame untuk visualisasi
subset_df = tfidf_df.iloc[:, :10]  # Mengambil 10 kolom pertama sebagai contoh

plt.figure(figsize=(12, 8))
sns.heatmap(subset_df, cmap='YlGnBu', annot=True, fmt=".2f")
plt.title('Heatmap TF-IDF')
plt.xlabel('Kata')
plt.ylabel('Dokumen')
plt.show()
```

```python
from wordcloud import WordCloud

# Menggabungkan semua nilai TF-IDF untuk membuat word cloud
wordcloud_data = tfidf_df.sum(axis=0)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(wordcloud_data)

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud Berdasarkan TF-IDF')
plt.show()
```

```python
# Mengambil kata-kata tertinggi dari dokumen pertama
top_n = 10
top_words = tfidf_df.iloc[0].nlargest(top_n)

plt.figure(figsize=(10, 6))
top_words.plot(kind='barh')
plt.title('Top 10 Kata Teratas di Dokumen Pertama')
plt.xlabel('Nilai TF-IDF')
plt.ylabel('Kata')
plt.show()
```

```python
# Menggabungkan semua nilai TF-IDF untuk membuat word cloud
wordcloud_data = tfidf_df.sum(axis=0)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(wordcloud_data)

# Menyimpan word cloud ke file gambar
wordcloud.to_file('wordcloud_output.png')

# (Opsional) Menampilkan word cloud
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud Berdasarkan TF-IDF')
plt.show()
```
